# Sentiment Analysis from Hugging Face with especial model not defualt model

```python
from transformers import pipeline

# Load the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Text to analyze
text = "I love using Hugging Face models. They are very convenient and powerful."

# Analyze sentiment
result = sentiment_analyzer(text)

print(result)
```

```python
from transformers import pipeline

# Specify the model
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

# Load the sentiment analysis pipeline with the specific model
sentiment_analyzer = pipeline("sentiment-analysis", model=model_name)

# Example text
text = "This new coffee shop has the best espresso I've ever tasted!"

# Analyze sentiment
result = sentiment_analyzer(text)

print(result)
```
